```python
import re
# ファイル名から数字を抽出するための関数
def extract_number(file_path):
    match = re.search(r'_(\d+)\.txt$', file_path.name)
    return int(match.group(1)) if match else float('inf')
```

```python

def sort_test_list(txt_files):
    # ファイルを番号順にソート
    sorted_files = sorted(txt_files, key=extract_number)
    return sorted_files
```

```python
from pathlib import Path

# chunk_text_directory = "/content/drive/MyDrive/Colab Notebooks/chunk_text_1000"
# chunk_text_directory = "/content/drive/MyDrive/Colab Notebooks/chunk_text_20000"
chunk_text_directory = "/content/drive/MyDrive/Colab Notebooks/chunk_text_5000"

def get_text_list(text_path=chunk_text_directory):
    # テキストファイルの読み込み pathlib.Path
    text_path = Path(text_path)
    # read glob *.txt
    glob_list = list(text_path.glob("*.txt"))
    glob_list = sort_test_list(glob_list)
    return glob_list
```

```python

def read_text_file(file_path):
    with  open(file_path, 'r', encoding='utf-8',  errors='ignore') as f:
        text = f.read()
    return text
```

```python
def read_text_files(file_paths):
    if len(file_paths) == 0:
        yield ""
    for file_path in file_paths:
        text = read_text_file(file_path)
        yield text
```

```python
def tokenize_process(text):
    inputs = tokenizer(text, return_tensors="pt").to(device)
    # labels = inputs["input_ids"].to(device)
    with torch.no_grad():
        outputs = model(**inputs)

    # 最後のトークンの予測ロジットを取得
    next_token_logits = outputs.logits[:, -1, :]

    # ロジットをソフトマックスで確率に変換
    probabilities = torch.nn.functional.softmax(next_token_logits, dim=-1)

    top_k = 1
    top_one_probabilities, top_k_indices = torch.topk(probabilities, top_k)

    zeroth_token =0
    token = tokenizer.decode(top_k_indices[0, zeroth_token])

    text += token
    print(token)
    proba = probabilities[0].to("cpu")

    # メモリの開放
    del inputs, outputs, next_token_logits, probabilities, top_one_probabilities, top_k_indices, token
    torch.cuda.empty_cache()

    return proba, text
```

prob = torch.nn.functional.softmax(next_token_logits, dim=-1)

この行で、next_token_logitsにソフトマックス関数を適用した結果、新しいテンソルprobが生成されます。このテンソルはprobとして参照され、probabilities_listにappendされることになります。

テンソル生成の流れ
next_token_logits:

これは、outputs.logits[:, -1, :]によって生成されますが、outputs.logitsの一部をスライスして参照しているため、新しいテンソルとはみなされません。基本的には既存のテンソルの一部を指している状態です。
prob:

torch.nn.functional.softmax(next_token_logits, dim=-1)が適用された結果として、新しいテンソルが生成されます。probは、この新しく生成されたテンソルを指しています。

```python
import numpy as np

def dimension_reduction(pt, vocab_size):
    reduced_dim = 500
    """
    次元削減を行う関数

    :param pt: 各タイムステップでの語彙ベクトル (N x |V|)
    :param vocab_size: 元の語彙のサイズ |V|
    :param reduced_dim: 次元削減後の新しい次元数 M
    :return: 次元削減後のベクトル (N x M)
    """
    N = pt.shape[0]  # タイムステップの数
    reduced_pt = np.zeros((N, reduced_dim))  # 次元削減後のベクトル

    for t in range(N):
        for w in range(vocab_size):
            m = w % reduced_dim  # モジュロ関数でグループ化
            reduced_pt[t, m] += pt[t, w]  # グループ内の合計

    return reduced_pt

# # 使用例
# N = 10  # タイムステップの数
# vocab_size = 10000  # 語彙のサイズ
# reduced_dim = 1000  # 次元削減後の新しい次元数

# # ランダムなベクトルを生成
# pt = np.random.rand(N, vocab_size)

# # 次元削減を実行
# reduced_pt = dimension_reduction(pt, vocab_size, reduced_dim)

# print(reduced_pt)
```
